```python
def table_to_df(table_name, table_id, year, tract):
    """Get census tables for Washington state"""
    
    _dict = c.acs5.get(table_id,
                       year=year,
                       geo={'for': 'tract:'+tract,
                       'in': 'state:{}'.format(states.WA.fips)})
    df = pd.DataFrame.from_dict(_dict)
    df.rename(columns={table_id:'field_value'},inplace=True)
    df['field_name'] = table_name
    df['year'] = year
    
    return df
```

```python
# Dictionary of census tables
table_dict = {
    'white population': 'C02003_003E',
    'black or african american': 'C02003_004E',
    'american indian and alaska native': 'C02003_005E',
    'asian alone': 'C02003_006E'
}
```

```python
# Population by Race
df = pd.DataFrame()
for year in ['2009','2010','2011','2012','2013','2014','2015']:
    for table_name, table_id in table_dict.iteritems():
        df = df.append(table_to_df(table_name=table_name,
                                   year=year,
                                   table_id=table_id,
                                   tract='*'))

# add full tract ID
df['tractce'] = df['state']+df['county']+df['tract']
df.to_csv('R:/brice/displacement/census-data/race-acs.csv',index=False)
```

```python
# Population by Income
df = pd.DataFrame()
for year in ['2009','2010','2011','2012','2013','2014','2015']:
    df = df.append(table_to_df(table_name='median income',
                                   year=year,
                                   table_id='B06011_001E',
                                   tract='*'))

# add full tract ID
df['tractce'] = df['state']+df['county']+df['tract']
df.to_csv('R:/brice/displacement/census-data/median-income-acs.csv',index=False)
```
